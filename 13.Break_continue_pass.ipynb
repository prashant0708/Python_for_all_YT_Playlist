```python
# Real-World Example 1: User Login System

password = 'Admin123'

max_attempt = 3

for attempt in range (1,max_attempt+1 ): # 1,2,3
    user_password =  input("Enter your password ")
    if user_password == password:
        print("You have loged in Successfully")
        break
    else:
        print(f"Wrong password and you are left with attempt {max_attempt- attempt}")
else:
    print("Your account has been locked")
```

```python
## Real world example 2 , search product in the database

products = [
    {"id": 101, "name": "Laptop", "price": 50000},
    {"id": 102, "name": "Mouse", "price": 500},
    {"id": 103, "name": "Keyboard", "price": 1500},
    {"id": 104, "name": "Monitor", "price": 15000}
]

search_id = int(input ("enter the product id "))  #103

for product in products: 
    if search_id == product['id']:
        print(product)
        break
else:
    print("Product not Found")
```

```python
# Email validation system
emails = [
    "user1@example.com",
    "invalid-email",
    "user2@company.com",
    "",
    "user3@domain.org",
    "no-at-symbol.com"
]

for email in emails:
    if '@' not in email or len(email)<=0 :
        continue
    print(email)
```
