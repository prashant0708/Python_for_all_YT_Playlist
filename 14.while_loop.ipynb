## Definition

**A while loop is a control flow statement that repeats a block of code as long as a given condition is True.**

*The loop checks the condition before every iteration.*


### Real Life Concepts 
* Keep studying while exam is not over.       
* Keep filling water while the tank is not full.
* Keep riding bike untill you reach the destination.
* Keep Gas stove turn on untill Milk is not boiled 


### Key concept
* If the condition becomes False, the loop stops.
* If the condition never becomes False, it becomes an infinite loop.


### The while loop is important because it:
* allows repetition when the number of iterations is not known in advance.
* works based on conditions, not fixed ranges.
* is useful for real-time, dynamic, and input-based problems.


### Use a while loop when:
* You donâ€™t know how many times the loop should run.
* Loop depends on user input.
* Loop should continue until a condition changes.
* You need infinite looping with control.


### Important points:
* condition must return True or False
* Indentation is mandatory
* Condition update is very important to stop the loop

### Key Difference b/w while & for loop:

* Use for loop when the number of iterations is fixed.
* Use while loop when the number of iterations depends on a condition

| Situation            | Best Loop |
| -------------------- | --------- |
| Fixed range          | for       |
| Unknown iterations   | while     |
| User input           | while     |
| Infinite loop        | while     |
| List/tuple traversal | for       |

```python
## print even number from 1 to 20 using while loop 

i = 20

while i >=0:
    if i %2 ==0:
        print(f"Number is even {i}")
    i -= 1
        
else:
    print(f"out of loop because loop is completed {i}")
```

```python
## nested while condition 

i = 3

while i >=1:
    j = 1
    while j <= 3:
        print("*", end = " ")
        j+=1
        
    i -= 1
```
