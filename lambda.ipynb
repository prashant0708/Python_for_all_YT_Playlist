A Lambda function in Python is an anonymous function defined using the lambda keyword. 
It can take any number of arguments but only has one expression. 
It is often used when a short, simple function is required for a brief moment, and defining a full function using def would be overkill.

```python
## syntax
lambda arguments :expression
```

```python
def name_of_function
```

```python
## add any number to 10

add = lambda x : 10 +x

print(add(5))
```

```python
## add two number 
add = lambda a , b : a+b

print(add(10,60))
```

```python
## find number is even or odd

is_even = lambda a : "even" if a%2 ==0 else "odd"

print(is_even(56))
```

```python
for i in range (1,21):
    result=is_even(i)
    print(i,result)
```

```python
## square of two number 

square = lambda x : x*x
print(square(20))
```

```python
## find the simple interest using lambda function 

## s.i = (pxrxt)/100

simple_interest = lambda p ,r,t : (p*r*t)/100

print(simple_interest(50000,5,2))
```

```python
## find the maximum number amoung three number 

max_number = lambda a,b,c : a if(a>b and a>c) else (b if b>c else c)

print(max_number(2,20,5))
```

```python
## check the length of the password is valid or not
## if the length of passwrod > = 8 --> valid else --> invalid

check_pwd = lambda pwd :"valid" if len(pwd)>=8 else "invalid"

print(check_pwd("admin"))
```

```python
## short the student by marks 

students = [
    ("amit", 78), 
    ("Sourav", 99),
    ("Rohit", 45),
    ("Arti", 78)
    
]

students.sort(key=lambda x :x[1])
print(students)
```

```python
## marks grade system using lambda

grad = lambda x :  ("A" if x >=90 else "B" if x>=75 else  "C" if x>=50 else "Fail" )

print(grad(86))
```
