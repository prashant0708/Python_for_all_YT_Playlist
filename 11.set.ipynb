```python
### subset
## set A can be subset of B if all the elements of A coinside with B

a = {4,5,6}

b ={1,2,3,4,5,6}

a.issubset(b) 
```

```python
b.issubset(a)
```

```python
b.issuperset(a)
```

```python
a = {1,2,3}
b = {1,2,3}

a.issubset(b)
```

```python
b.issubset(a)
```

```python
b.issuperset(a)
```

```python
a.issuperset(b)
```
