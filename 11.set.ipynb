```python
## subset
## superset
## Disjoin


## subset


## A is sub set of B if All the value of A is exists in B 

a = {1,2,3,4,5,6}

b = {1,2,3,4,5,6,4,5,6,7,8}

a.issubset(b)
```

```python
b.issubset(a)
```

```python
### superset 

#if  A is subset of B it means B is superset

#if All the value of A is available in B it means A is subset of B and B is superset

b.issuperset(a)
```

```python
a.issuperset(b)
```

```python
## Disjoin

a.isdisjoint(b)
```

```python
a =  {1,2,3,4}
b = {6,7,8,9}
a.isdisjoint(b)
```
